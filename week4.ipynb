```python
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç  ùúí2ùëò

–ü–µ—Ä–≤—ã–π –±–ª–æ–∫ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —Ä–∞—Å–ø—Ä–µ–¥–µ–Ω–∏–µ ùúí2ùëò.
–ú—ã –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º 100 —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ —ç—Ç–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é. –ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∞ —á–∞—Å—Ç–æ—Ç–∞ –∫–∞–∂–¥–æ–≥–æ —Å–ª—É—á–∞–π–Ω–æ–≥–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–æ–≤–µ—Ä—Ö —Å–∏–Ω–µ–π –ª–∏–Ω–∏–µ–π.
–°–æ–≥–ª–∞—Å–Ω–æ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º –¥–∞–Ω–Ω—ã–º, –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Ç–∞–∫–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–≤–Ω–æ ùëò, –∞ –¥–∏—Å–ø–µ—Ä—Å–∏—è 2ùëò.

```python
n = 1000
x = np.linspace(0,30,n)
k = 3

rv = sts.chi2(k)
y = rv.pdf(x)
plt.plot(x, y, label="–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏")
    
sample = rv.rvs(n)
plt.hist(sample, density=True, bins=100)
plt.ylabel('$F(x)$')
plt.xlabel('$x$') 

plt.legend()
```

# –ò–ª–ª—é—Å—Ç–∞—Ü–∏—è –¶–ü–¢

## –í—ã–±–æ—Ä–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ n = 5

–í –¥–∞–Ω–Ω–æ–º –±–ª–æ–∫–µ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç—Å—è 1000 —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –≤—ã–±–æ—Ä–æ–∫ –ø–æ 5 –∑–Ω–∞—á–µ–Ω–∏–π –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è ùúí2ùëò. –í –∫–∞–∂–¥–æ–π –≤—ã–±–æ—Ä–∫–µ —Å—á–∏—Ç–∞–µ—Ç—Å—è —Å—Ä–µ–¥–µ–µ, –∞ –∑–∞—Ç–µ–º —Å—Ç—Ä–æ–∏—Ç—Å—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞, –Ω–∞ –æ—Å–Ω–æ–Ω–µ 1000 –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö. –ü–æ–≤–µ—Ä—Ö –≥–∏—Å—Ç–æ–≥—Ä–∞–º—ã –Ω–∞–ª–æ–∂–µ–Ω–∞ –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ —Å—Ä–µ–¥–Ω–∏–º —Ä–∞–≤–Ω—ã–º E = ùëò –∏ –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π <tex>$D = \frac{2ùëò}{n}$</tex>, –≥–¥–µ n = 5 (–≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –¶–ü–¢). –¢–∞–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤–∑—è—Ç—ã –∏–∑ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö –æ ùúí2ùëò —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –¥–ª—è –ø–æ—Å—Ç–Ω–æ–µ–Ω–∏—è –Ω–æ—Ä–º–∞–ª—å–Ω–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω—É–∂–µ–Ω –∏–º–µ–Ω–Ω–æ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª, –∏ –ø–æ —ç—Ç–æ–π –ø—Ä–∏—á–∏–Ω–µ –±—ã–ª –≤–∑—è—Ç –∫–æ—Ä–µ–Ω—å.

```python
means = []
for _ in range(1000):
    n = 5
    x = np.linspace(0,30,n)
    k = 3

    rv = sts.chi2(k)
    y = rv.pdf(x)
    sample = rv.rvs(n)

    means.append(float(sample.mean()))

means = np.array(means)
x = np.linspace(-5,10,100)
norm_rv = sts.norm(loc=k, scale=float(np.sqrt((2*k)/n)))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label = '–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏')

plt.hist(means, density=True, bins=100)
plt.ylabel('$F(x)$')
plt.xlabel('$x$') 

plt.legend()
```
